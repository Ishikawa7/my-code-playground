```python
### funzione per l'ordinamento e l'unione dei sotto insiemi (FUNZIONE 1)
def merge(arr, left, mid, right):
    print("FUNZIONE MERGE")
    print("ARRAY passato a merge: "+str([arr[i] for i in range(left, right+1)]))
    print("Indice left {}, indice mid {}, indice right {}: ".format(left, mid, right))
    dim_left = mid - left + 1
    dim_right = right - mid 
    
    left_arr = []
    right_arr = []

    for i in range(left, mid+1):
        left_arr.append(arr[i])
    
    for i in range(mid + 1, right+1):
        right_arr.append(arr[i])

    print("Array ausiliario left: " + str(left_arr))
    print("Array ausiliario right: " + str(right_arr))

    i_left = 0
    i_right = 0
    i_merge = left

    while i_left < dim_left and i_right < dim_right:
        print("ARRAY in fase di ordinamento: "+str([arr[i] for i in range(left, right+1)]))
        print("Indice left ausiliario {}, indice merge {}, indice right ausiliario {}: ".format(i_left, i_right, i_merge))
        if left_arr[i_left] < right_arr[i_right]:
            arr[i_merge] = left_arr[i_left]
            i_left += 1
        else:
            arr[i_merge] = right_arr[i_right]
            i_right += 1

        i_merge += 1

    while i_left < dim_left:
        arr[i_merge] = left_arr[i_left]
        i_left += 1
        i_merge += 1
        print("ARRAY in fase di ordinamento: "+str([arr[i] for i in range(left, right+1)]))

    while i_right < dim_right:
        arr[i_merge] = right_arr[i_right]
        i_right += 1
        i_merge += 1
        print("ARRAY in fase di ordinamento: "+str([arr[i] for i in range(left, right+1)]))
```

```python
### funzione che effettua la divisione in sottoparti e la ricorsione chiamando la FUNZIONE 1
def mergeSort(arr, left, right):
    print("FUNZIONE MERGE SORT")
    print("ARRAY passato a merge sort: "+str([arr[i] for i in range(left, right+1)]))
    if left < right:
        mid = (left + right) // 2 # - left serve per considerare quando left non e' zero

        print("Indice left {}, indice mid {}, indice right {}: ".format(left, mid, right))
        print("Parte sinistra: " + str([arr[i] for i in range(left, mid+1)]))
        print("Parte destra: " + str([arr[i] for i in range(mid+1, right+1)]))
        mergeSort(arr, left, mid)
        mergeSort(arr, mid+1, right)
        merge(arr, left, mid, right)
```

```python
arr = [3,1,2,4,3,6]

mergeSort(arr, 0, len(arr)-1)

print("ARRAY ordinato: "+str(arr))
```
